```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter
import seaborn as sns

class AIEconomicModel:
    def __init__(self, sectors=3, time_periods=20):
        # Initialize the model parameters
        self.n_sectors = sectors
        self.time_periods = time_periods
        self.setup_parameters()
        self.initialize_variables()

    def setup_parameters(self):
        """Set up the model parameters based on the paper's framework"""
        # Exogenous parameters (for baseline scenario)

        # Sectoral TFP
        self.A = np.array([1.0, 1.2, 0.9])  # TFP for each sector

        # Labor force (fixed)
        self.NH = 50  # High-skilled workers
        self.NL = 100  # Low-skilled workers

        # Production function elasticities
        self.alpha = np.array([0.3, 0.25, 0.35])  # Capital elasticity
        self.beta_H = np.array([0.3, 0.4, 0.15])  # High-skill labor elasticity
        self.beta_L = np.array([0.3, 0.2, 0.4])   # Low-skill labor elasticity
        self.gamma = np.array([0.1, 0.15, 0.08])  # AI capital elasticity

        # AI investment parameters
        self.a = np.array([0.5, 0.8, 0.3])  # Autonomous component
        self.lambda_param = np.array([0.2, 0.3, 0.1])  # Responsiveness to rental rate

        # Policy parameters
        self.tau = 0.0  # Tax rate (initially zero)
        self.theta = 0.0  # Retraining effectiveness (initially zero)

        # Prices (normalized to 1 for simplicity)
        self.P = np.ones(self.n_sectors)

    def initialize_variables(self):
        """Initialize the state variables for the simulation"""
        # Arrays to store time series for all variables
        self.Y = np.zeros((self.time_periods, self.n_sectors))  # Output by sector
        self.K = np.zeros((self.time_periods, self.n_sectors))  # Physical capital by sector
        self.AI = np.zeros((self.time_periods, self.n_sectors))  # AI capital by sector
        self.LH = np.zeros((self.time_periods, self.n_sectors))  # High-skill labor by sector
        self.LL = np.zeros((self.time_periods, self.n_sectors))  # Low-skill labor by sector
        self.wH = np.zeros(self.time_periods)  # High-skill wage
        self.wL = np.zeros(self.time_periods)  # Low-skill wage
        self.r = np.zeros(self.time_periods)   # Rental rate
        self.UH = np.zeros(self.time_periods)  # High-skill unemployment
        self.UL = np.zeros(self.time_periods)  # Low-skill unemployment
        self.UBI = np.zeros(self.time_periods) # UBI payment per person
        self.I_AI = np.zeros((self.time_periods, self.n_sectors))  # AI investment by sector

        # Initial values
        self.K[0] = np.array([50.0, 40.0, 30.0])
        self.AI[0] = np.array([5.0, 8.0, 3.0])
        self.LH[0] = np.array([20.0, 25.0, 5.0])
        self.LL[0] = np.array([40.0, 20.0, 40.0])
        self.r[0] = 0.1

        # Calculate initial output
        for i in range(self.n_sectors):
            self.Y[0, i] = self.production_function(
                self.A[i], self.K[0, i], self.LH[0, i], self.LL[0, i], self.AI[0, i],
                self.alpha[i], self.beta_H[i], self.beta_L[i], self.gamma[i]
            )

        # Initial wages based on marginal products
        self.wH[0] = np.mean([self.P[i] * (self.beta_H[i] * self.Y[0, i] / self.LH[0, i]) for i in range(self.n_sectors)])
        self.wL[0] = np.mean([self.P[i] * (self.beta_L[i] * self.Y[0, i] / self.LL[0, i]) for i in range(self.n_sectors)])

        # Initial unemployment rates
        self.UH[0] = max(0, (self.NH - np.sum(self.LH[0])) / self.NH)
        self.UL[0] = max(0, (self.NL - np.sum(self.LL[0])) / self.NL)

    def production_function(self, A, K, LH, LL, AI, alpha, beta_H, beta_L, gamma):
        """Cobb-Douglas production function for a sector"""
        return A * (K ** alpha) * ((LH ** beta_H) * (LL ** beta_L)) * (AI ** gamma)

    def update_AI_investment(self, t):
        """Update AI investment based on rental rate"""
        for i in range(self.n_sectors):
            self.I_AI[t, i] = self.a[i] + self.lambda_param[i] * self.r[t-1]

    def update_capital_stocks(self, t):
        """Update physical and AI capital stocks"""
        # AI capital accumulation
        for i in range(self.n_sectors):
            self.AI[t, i] = self.AI[t-1, i] + self.I_AI[t-1, i]

        # Physical capital is assumed to grow exogenously for simplicity
        for i in range(self.n_sectors):
            # Capital growth is influenced by output
            growth_rate = 0.05 * (1 + 0.2 * (self.Y[t-1, i] / self.Y[0, i] - 1))
            self.K[t, i] = self.K[t-1, i] * (1 + growth_rate)

    def allocate_labor(self, t):
        """Allocate labor across sectors based on wages"""
        # Calculate labor demand based on marginal products
        LH_demand = np.zeros(self.n_sectors)
        LL_demand = np.zeros(self.n_sectors)

        for i in range(self.n_sectors):
            # Labor demand derived from profit maximization
            LH_demand[i] = (self.P[i] * self.beta_H[i] * self.Y[t-1, i]) / self.wH[t-1]
            LL_demand[i] = (self.P[i] * self.beta_L[i] * self.Y[t-1, i]) / self.wL[t-1]

        # Apply retraining effects (if any)
        if t > 0 and self.theta > 0:
            # Some unemployed low-skilled workers become high-skilled
            retrained_workers = self.theta * self.UL[t-1] * self.NL
            effective_NH = self.NH + retrained_workers
            effective_NL = self.NL - retrained_workers
        else:
            effective_NH = self.NH
            effective_NL = self.NL

        # Normalize demands to respect labor force constraints
        total_LH_demand = np.sum(LH_demand)
        total_LL_demand = np.sum(LL_demand)

        if total_LH_demand > effective_NH:
            LH_demand = LH_demand * (effective_NH / total_LH_demand)

        if total_LL_demand > effective_NL:
            LL_demand = LL_demand * (effective_NL / total_LL_demand)

        # Update labor allocations
        self.LH[t] = LH_demand
        self.LL[t] = LL_demand

        # Update unemployment rates
        self.UH[t] = max(0, (effective_NH - np.sum(self.LH[t])) / effective_NH)
        self.UL[t] = max(0, (effective_NL - np.sum(self.LL[t])) / effective_NL)

    def calculate_output(self, t):
        """Calculate sectoral and aggregate output"""
        for i in range(self.n_sectors):
            self.Y[t, i] = self.production_function(
                self.A[i], self.K[t, i], self.LH[t, i], self.LL[t, i], self.AI[t, i],
                self.alpha[i], self.beta_H[i], self.beta_L[i], self.gamma[i]
            )

    def update_prices(self, t):
        """Update factor prices (wages and rental rate)"""
        # Calculate marginal products
        MPH = np.array([self.P[i] * (self.beta_H[i] * self.Y[t, i] / self.LH[t, i]) for i in range(self.n_sectors)])
        MPL = np.array([self.P[i] * (self.beta_L[i] * self.Y[t, i] / self.LL[t, i]) for i in range(self.n_sectors)])
        MPK = np.array([self.P[i] * (self.alpha[i] * self.Y[t, i] / self.K[t, i]) for i in range(self.n_sectors)])

        # Update wages and rental rate (weighted by sectoral output)
        weights = self.Y[t] / np.sum(self.Y[t])
        self.wH[t] = np.sum(MPH * weights)
        self.wL[t] = np.sum(MPL * weights)
        self.r[t] = np.sum(MPK * weights)

    def calculate_UBI(self, t):
        """Calculate UBI payment if tax rate is positive"""
        if self.tau > 0:
            total_output = np.sum(self.Y[t])
            tax_revenue = self.tau * total_output
            self.UBI[t] = tax_revenue / (self.NH + self.NL)
        else:
            self.UBI[t] = 0

    def run_simulation(self):
        """Run the simulation for the specified time periods"""
        for t in range(1, self.time_periods):
            # Update AI investment
            self.update_AI_investment(t)

            # Update capital stocks
            self.update_capital_stocks(t)

            # Allocate labor across sectors
            self.allocate_labor(t)

            # Calculate output
            self.calculate_output(t)

            # Update prices
            self.update_prices(t)

            # Calculate UBI
            self.calculate_UBI(t)

    def run_scenario(self, scenario_name, **params):
        """Run a specific scenario by modifying parameters"""
        # Save original parameters
        original_params = {
            'gamma': self.gamma.copy(),
            'tau': self.tau,
            'theta': self.theta,
            'lambda_param': self.lambda_param.copy()
        }

        # Update parameters for this scenario
        for param, value in params.items():
            setattr(self, param, value)

        # Re-initialize and run simulation
        self.initialize_variables()
        self.run_simulation()

        # Collect results
        results = {
            'scenario': scenario_name,
            'Y_total': np.sum(self.Y, axis=1),
            'Y_sectors': self.Y.copy(),
            'wH': self.wH.copy(),
            'wL': self.wL.copy(),
            'wRatio': self.wH / self.wL,
            'UH': self.UH.copy(),
            'UL': self.UL.copy(),
            'U_total': (self.UH * self.NH + self.UL * self.NL) / (self.NH + self.NL),
            'AI_total': np.sum(self.AI, axis=1),
            'UBI': self.UBI.copy()
        }

        # Restore original parameters
        for param, value in original_params.items():
            setattr(self, param, value)

        return results

# Function to run multiple scenarios and visualize results
def run_multiple_scenarios():
    # Initialize model
    model = AIEconomicModel(sectors=3, time_periods=30)

    # Define scenarios
    scenarios = [
        # Baseline scenario
        {
            'name': 'Baseline',
            'params': {}
        },
        # Accelerated AI adoption
        {
            'name': 'Accelerated AI Adoption',
            'params': {
                'gamma': np.array([0.2, 0.3, 0.15]),
                'lambda_param': np.array([0.4, 0.6, 0.2])
            }
        },
        # UBI policy
        {
            'name': 'UBI Policy',
            'params': {
                'tau': 0.1,
                'gamma': np.array([0.15, 0.25, 0.12])
            }
        },
        # Retraining program
        {
            'name': 'Retraining Program',
            'params': {
                'theta': 0.2,
                'gamma': np.array([0.15, 0.25, 0.12])
            }
        },
        # Combined policies
        {
            'name': 'Combined Policies',
            'params': {
                'tau': 0.05,
                'theta': 0.15,
                'gamma': np.array([0.15, 0.25, 0.12])
            }
        }
    ]

    # Run all scenarios
    results = []
    for scenario in scenarios:
        print(f"Running scenario: {scenario['name']}")
        scenario_results = model.run_scenario(scenario['name'], **scenario['params'])
        results.append(scenario_results)

    return results

# Visualization functions
def plot_output_growth(results):
    plt.figure(figsize=(12, 6))

    for result in results:
        # Normalize output to initial period
        normalized_output = result['Y_total'] / result['Y_total'][0]
        plt.plot(normalized_output, label=result['scenario'])

    plt.title('Economic Growth Under Different Scenarios', fontsize=16)
    plt.xlabel('Time Periods', fontsize=14)
    plt.ylabel('Output Index (Period 0 = 1)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()

    return plt.gcf()

def plot_wage_inequality(results):
    plt.figure(figsize=(12, 6))

    for result in results:
        plt.plot(result['wRatio'], label=result['scenario'])

    plt.title('Wage Inequality (High-Skill/Low-Skill Wage Ratio)', fontsize=16)
    plt.xlabel('Time Periods', fontsize=14)
    plt.ylabel('Wage Ratio', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()

    return plt.gcf()

def plot_unemployment(results):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    for result in results:
        ax1.plot(result['UL'] * 100, label=result['scenario'])
        ax2.plot(result['UH'] * 100, label=result['scenario'])

    ax1.set_title('Low-Skill Unemployment Rate', fontsize=16)
    ax1.set_xlabel('Time Periods', fontsize=14)
    ax1.set_ylabel('Unemployment Rate (%)', fontsize=14)
    ax1.grid(True, linestyle='--', alpha=0.7)

    ax2.set_title('High-Skill Unemployment Rate', fontsize=16)
    ax2.set_xlabel('Time Periods', fontsize=14)
    ax2.set_ylabel('Unemployment Rate (%)', fontsize=14)
    ax2.grid(True, linestyle='--', alpha=0.7)

    ax1.legend()
    ax2.legend()
    plt.tight_layout()

    return fig

def plot_ai_adoption(results):
    plt.figure(figsize=(12, 6))

    for result in results:
        plt.plot(result['AI_total'], label=result['scenario'])

    plt.title('AI Capital Accumulation Across Scenarios', fontsize=16)
    plt.xlabel('Time Periods', fontsize=14)
    plt.ylabel('Total AI Capital', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()

    return plt.gcf()

def plot_sector_impact(results, scenario_index):
    scenario = results[scenario_index]

    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    sectors = ['Manufacturing', 'Services', 'Agriculture']

    for i in range(3):
        # Normalize output for easier comparison
        normalized_output = scenario['Y_sectors'][:, i] / scenario['Y_sectors'][0, i]
        axs[i].plot(normalized_output, 'b-', linewidth=2)
        axs[i].set_title(f'Sector {i+1}: {sectors[i]}', fontsize=16)
        axs[i].set_xlabel('Time Periods', fontsize=14)
        axs[i].set_ylabel('Output Index (Period 0 = 1)', fontsize=14)
        axs[i].grid(True, linestyle='--', alpha=0.7)

    plt.suptitle(f'Sectoral Impact - {scenario["scenario"]} Scenario', fontsize=18)
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)

    return fig

def plot_policy_comparison(results):
    # Extract the relevant scenarios (UBI, Retraining, Combined)
    policy_results = [r for r in results if r['scenario'] in ['Baseline', 'UBI Policy', 'Retraining Program', 'Combined Policies']]

    fig, axs = plt.subplots(2, 2, figsize=(16, 12))

    # Output Growth
    for result in policy_results:
        normalized_output = result['Y_total'] / result['Y_total'][0]
        axs[0, 0].plot(normalized_output, label=result['scenario'])

    axs[0, 0].set_title('Economic Growth', fontsize=16)
    axs[0, 0].set_xlabel('Time Periods', fontsize=14)
    axs[0, 0].set_ylabel('Output Index (Period 0 = 1)', fontsize=14)
    axs[0, 0].grid(True, linestyle='--', alpha=0.7)
    axs[0, 0].legend()

    # Wage Inequality
    for result in policy_results:
        axs[0, 1].plot(result['wRatio'], label=result['scenario'])

    axs[0, 1].set_title('Wage Inequality', fontsize=16)
    axs[0, 1].set_xlabel('Time Periods', fontsize=14)
    axs[0, 1].set_ylabel('High-Skill/Low-Skill Wage Ratio', fontsize=14)
    axs[0, 1].grid(True, linestyle='--', alpha=0.7)
    axs[0, 1].legend()

    # Low-Skill Unemployment
    for result in policy_results:
        axs[1, 0].plot(result['UL'] * 100, label=result['scenario'])

    axs[1, 0].set_title('Low-Skill Unemployment Rate', fontsize=16)
    axs[1, 0].set_xlabel('Time Periods', fontsize=14)
    axs[1, 0].set_ylabel('Unemployment Rate (%)', fontsize=14)
    axs[1, 0].grid(True, linestyle='--', alpha=0.7)
    axs[1, 0].legend()

    # UBI Payments
    for result in policy_results:
        axs[1, 1].plot(result['UBI'], label=result['scenario'])

    axs[1, 1].set_title('UBI Payments', fontsize=16)
    axs[1, 1].set_xlabel('Time Periods', fontsize=14)
    axs[1, 1].set_ylabel('UBI per Person', fontsize=14)
    axs[1, 1].grid(True, linestyle='--', alpha=0.7)
    axs[1, 1].legend()

    plt.suptitle('Policy Impact Comparison', fontsize=20)
    plt.tight_layout()
    plt.subplots_adjust(top=0.92)

    return fig

def create_summary_dashboard(results):
    """Create a dashboard with key summary metrics for all scenarios"""
    # Create a dictionary to hold final period values for key metrics
    summary_data = {}
    metrics = ['Output Growth (%)', 'Wage Inequality Ratio', 'Low-Skill Unemployment (%)',
               'High-Skill Unemployment (%)', 'Total AI Capital', 'UBI Payment']

    for result in results:
        # Calculate values for final period (relative to initial period for growth)
        output_growth = ((result['Y_total'][-1] / result['Y_total'][0]) - 1) * 100
        wage_inequality = result['wRatio'][-1]
        low_skill_unemp = result['UL'][-1] * 100
        high_skill_unemp = result['UH'][-1] * 100
        ai_capital = result['AI_total'][-1]
        ubi = result['UBI'][-1]

        # Store in dictionary
        summary_data[result['scenario']] = [
            output_growth, wage_inequality, low_skill_unemp,
            high_skill_unemp, ai_capital, ubi
        ]

    # Convert to DataFrame
    df = pd.DataFrame(summary_data, index=metrics)

    # Create heatmap for visual comparison
    plt.figure(figsize=(14, 8))
    sns.heatmap(df, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=0.5)
    plt.title('Scenario Comparison - Final Period Metrics', fontsize=18)
    plt.tight_layout()

    return plt.gcf(), df

# Main function to run the simulation and create visualizations
def main():
    # Run simulations
    results = run_multiple_scenarios()

    # Create visualization figures
    fig1 = plot_output_growth(results)
    fig2 = plot_wage_inequality(results)
    fig3 = plot_unemployment(results)
    fig4 = plot_ai_adoption(results)
    fig5 = plot_sector_impact(results, 1)  # Accelerated AI Adoption scenario
    fig6 = plot_policy_comparison(results)
    fig7, summary_df = create_summary_dashboard(results)

    # Display summary dataframe
    print("\nSummary of Results (Final Period):")
    print(summary_df)

    # Save figures
    fig1.savefig('output_growth.png')
    fig2.savefig('wage_inequality.png')
    fig3.savefig('unemployment.png')
    fig4.savefig('ai_adoption.png')
    fig5.savefig('sector_impact.png')
    fig6.savefig('policy_comparison.png')
    fig7.savefig('summary_dashboard.png')

    print("\nSimulation complete. Figures saved.")

    # Show the figures (uncomment if running in interactive environment)
    plt.show()

if __name__ == "__main__":
    main()
```
